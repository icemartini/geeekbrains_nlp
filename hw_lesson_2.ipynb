1. Создайте мешок слов с помощью sklearn.feature_extraction.text.CountVectorizer.fit_transform(). Применим его к 'tweet_stemmed' и 'tweet_lemmatized' отдельно.

Проверьте ваши векторайзеры на корпусе который использовали на вебинаре, составьте таблицу метод векторизации и скор который вы получили (в методах векторизации по изменяйте параметры что бы добиться лучшего скора) обратите внимание как падает/растёт скор при уменьшении количества фичей, и изменении параметров, так же попробуйте применить к векторайзерам PCA для сокращения размерности посмотрите на качество сделайте выводы

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'], random_state=24)

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)

count_vectorizer.fit(trainDF['text'])

xtrain_count = count_vectorizer.transform(train_x)
xvalid_count = count_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
pred_count = classifier.predict(xvalid_count)
pred_count
```

```python
tfidf_vectorizer.fit(trainDF['text'])

xtrain_tfidf = tfidf_vectorizer.transform(train_x)
xvalid_tfidf = tfidf_vectorizer.transform(valid_x)

classifier.fit(xtrain_tfidf, train_y)
pred_tfidf = classifier.predict(xvalid_tfidf)
pred_tfidf
```

```python
d = {'count_vectorizer': accuracy_score(valid_y, pred_count), 'tfidf_vectorizer': accuracy_score(valid_y, pred_tfidf)}
acc_df = pd.DataFrame(data=d, index=['accuracy'])
acc_df
```

```python
for n in range(1,11):
    count_vect = CountVectorizer(max_df = 0.9, max_features = n*500, stop_words='english')
    tfidf_vect = TfidfVectorizer(max_df = 0.9, max_features = n*500, stop_words='english')

    count_vect.fit(trainDF['text'])
    tfidf_vect.fit(trainDF['text'])

    xtrain_count1 = count_vect.transform(train_x)
    xvalid_count1 = count_vect.transform(valid_x)
    xtrain_tfidf1 = tfidf_vect.transform(train_x)
    xvalid_tfidf1 = tfidf_vect.transform(valid_x)

    classifier.fit(xtrain_count1, train_y)
    pred_count1 = classifier.predict(xvalid_count1)
    
    classifier.fit(xtrain_tfidf1, train_y)
    pred_tfidf1 = classifier.predict(xvalid_tfidf1)

    print(f'max_features: {n*500}, \
            count_accuracy: {accuracy_score(valid_y, pred_count1)}, \
            tfidf_accuracy: {accuracy_score(valid_y, pred_tfidf1)}')
```

6. Поскольку наши данные содержат твиты, а не только слова, нам придется придумать способ использовать векторы слов из модели word2vec для создания векторного представления всего твита. Существует простое решение этой проблемы, мы можем просто взять среднее значение всех векторов слов, присутствующих в твите. Длина результирующего вектора будет одинаковой, то есть 200. Мы повторим тот же процесс для всех твитов в наших данных и получим их векторы. Теперь у нас есть 200 функций word2vec для наших данных.

Необходимо создать вектор для каждого твита, взяв среднее значение векторов слов, присутствующих в твите. В цикле сделать:  vec += model_w2v[word].reshape((1, size))
и поделить финальный вектор на количество слов в твите.
На выходе должен получиться wordvec_df.shape = (49159, 200).


```python
wordvec_df = np.zeros((len(df), 200))
for i, tweet in enumerate(df['tweet_token']):
    vec = np.zeros((1, 200))
    for word in tweet:
        try:
            vec += modelW2V.wv[word].reshape((1, 200))
        except:
            continue
    vec /= len(tweet)
    wordvec_df[i] = vec
```
